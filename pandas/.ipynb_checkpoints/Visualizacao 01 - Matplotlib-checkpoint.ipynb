```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(1, 10, 100)
x
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x), 'r')
```

```python
plt.plot(x, np.sin(x), '--')
```

```python
plt.plot(x, np.sin(x), 'r--')
```
