```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario = np.array([100, 200, 300, 500, 400, 150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='black', ls='--', marker='s', ms=15)
plt.show()
```

```python
salarios_marcos = np.array([100, 200, 400, 300])
salarios_gileno = np.array([50, 300, 500, 450])
```

```python
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salários Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=8, label='Salários Gileno')
plt.legend()
plt.show()
```

```python
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=8, label='Salários Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='s', ms=8, label='Salários Gileno')
plt.legend(loc='upper right')
plt.show()
```

```python
salarios_pedro = [200, 150, 300, 700]
plt.plot(salarios_marcos, c='Black', ls='--', marker='s', ms=10, label='Salários Marcos')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=10, label='Salários Gileno')
plt.plot(salarios_pedro, c='Green', ls='--', marker='o', ms=10, label='Salários Pedro')
plt.legend()
plt.show()
```
